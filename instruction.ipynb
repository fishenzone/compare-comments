## Описание сервиса

Сервис выполняет анализ документов и связанных с ними комментариев, позволяя:
- Загружать две версии документа (исходную и измененную)
- Загружать CSV-файл с комментариями к первой версии документа
- Автоматически определять, были ли учтены комментарии во второй версии документа
- Формировать подробный отчет о результатах анализа
